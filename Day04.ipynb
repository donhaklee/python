```python
# IF 문
# 조건문 - if 문
# 조건절에 0. False값 이외의 값이 있으면 참 
# 코드의 실행을 제어하기 위한 제어문 중 하나
# 조건식에 따라 코드의 실행을 제어한다.
# if 문에서 조건식의 결과는 항상 True, False로 판단을 한다.
# 조건식의 결과(True, False)에 따라 코드의 실행을 제어
 
# 사용 형식(Syntax)
# if 조건식:
#     실행 코드 1
# else:
#     실행 코드 2
#     ...
# 주의 사항
#   - if 문 끝에는 항상 :(콜론)이 있어야 한다.
#   - if 조건이 참(True)에 해당하는 경우 실행 할 코드는 반드시
#     들여쓰기(indent)를 적용해야 한다. indent는 공백 4칸 사용
#   - indent(들여쓰기)는 항상 동일한 공백 4칸을 사용해야 한다.
 
# if ... elif ... else
# if 조건식에 만족하지 않은 경우 또 다른 조건식을 평가하여
# 코드를 분기하기 위한 방법으로 사용
# elif는 반복적으로 사용 가능하다.
# else는 생략 가능하다.
 
# 사용 형식(Syntax)
# if 조건식1:
#     실행 코드 1
#     ...
# elif 조건식2:
#     실행 코드 2
#     ...
# else:
#     실행 코드 3
    …
```

```python
# if문은 조건부 실행에 실행된다.
grade = int(input('Enter score: '))
if grade >= 95 : 
  print('A+')
elif grade >= 90 : 
  print('A-')
elif grade >= 85 : 
  print('B+')
elif grade >= 80 : 
  print('B-')
elif grade >= 75 : 
  print('C+')
elif grade >= 70 : 
  print('C-')
elif grade >= 65 : 
  print('D+')
elif grade >= 60 : 
  print('D-')
else: print('F')
```

```python
# - 사용자로 부터 Gbyte의 값을 입력받아 byte,Kbyte,Mbyte로 각각 출력 되게 만드시오.
# - (1.byte 2.Kbyte 3.Mbyte 선택)
GB = int(input("Gbyte의 값을 입력하세요: "))
num = int(input("1.byte 2.Kbyte 3.Mbyte 선택 : "))
if num==1 :
    print("byte값은 : {:,} bytes".format(GB*1024**3))
elif num==2 : 
    print("Kbyte값은 : {:,} Kbytes".format(GB*1024**2))
elif num==3 : 
    print("Mbyte값은 : {:,} Mbytes".format(GB*1024))
else :
    print("메뉴 선택이 잘 못됐습니다.")
```

```python
# 10 ~ 99 까지 임의의 값 2개를 생성하고, 빼기 계산을 할 때 항상
# 양의 정수가 나올 수 있도록 하시오.
from random import randint
rand1 = randint(10, 99)
rand2 = randint(10, 99)
 
if rand1 < rand2:
    rand1, rand2 = rand2, rand1
print('{} - {} = {}'.format(rand1, rand2, rand1 - rand2))
```

```python
# 10 ~ 99 까지 임의의 값 2개를 생성하고 홀/짝을 비교하였을때 
# 2개의 값이 전부 홀수 또는 짝수이면, 2개의 정수 값을 더하고,
# 2개의 값이 홀-짝 또는 짝-홀이면, 2개의 정수 값을 곱하시오.
from random import randint
rand1 = randint(10, 99)
rand2 = randint(10, 99)
 
if rand1 % 2 == 0 and rand2 % 2 == 0:
    print('{} + {} = {}'.format(rand1, rand2, rand1 + rand2))
elif rand1 % 2 == 1 and rand2 % 2 == 1:
    print('{} + {} = {}'.format(rand1, rand2, rand1 + rand2))
else:
    print('{} * {} = {}'.format(rand1, rand2, rand1 * rand2))
```

```python
# 입력받은 숫자의 홀짝을 가리는 코드입니다.
# try절이 먼저 실행되고, 에러가 발생하면 except가 실행된다.
try:
    num = int(input('Enter the numbers here: '))
    odd = f'{num} is the odd.'
    even = f'{num} is the even.'    
except:
    print('Please, Enter the numbers only')
else:
    print(even) if not num % 2 else print(odd)
finally:
    print('Thank for all.')
```
